```r
library(dplyr)
library(ggplot2)

wd <- "C:/Users/saavedrak/Desktop/USP_587_data/HW3/"
df.demog <- read.csv(paste(wd, "hw3_demog.csv", sep=""))
df.trips.inc <- read.csv(paste(wd, "hw3_trip_rates.csv", sep=""))
df.trips.hhsiz <- read.csv(paste(wd, "hw3_trip_rates_hhsiz.csv", sep=""))
df.trips.obs <- read.csv(paste(wd, "hw3_observed.csv", sep=""))
```

```r
df.mult <- df.trips.hhsiz[1:5, c('hhsiz','htrips')] #extract multiplier
df.mult[1,]
```

```r
df.prod.attr <- df.demog %>%
    select(TAZ, hh1, hh2, hh3, hh4, hh5, empl_tot) %>%
    mutate(hh1 = hh1 * 3.7 *.2) %>%
    mutate(hh2 = hh2 * 7.6 * .22) %>%
    mutate(hh3 = hh3 * 10.6 * .19) %>%
    mutate(hh4 = hh4 * 13.6 * .19) %>%
    mutate(hh5 = hh5 * 16.6 * .17) %>%
    mutate(hbw_prod = hh1 + hh2 + hh3 + hh4) %>%
    mutate(hbw_attract = empl_tot * 1.45) %>%
    mutate(bal_factor = (sum(hbw_prod) / sum(hbw_attract))) %>%
    mutate(A_prime = bal_factor * hbw_attract)


df.prod.attr
```
